# LOAD

```python
import pandas as pd

# Replace 'your_dataset.json' with the actual file path
train = 'train.json'
test = 'test.json'

# Read JSON file and create a DataFrame
df_train = pd.read_json(train)
df_test = pd.read_json(test)
```

```python
X_train = df_train.drop("pret", axis=1)
y_train = df_train["pret"]
```

```python
X_test = df_test.drop("pret", axis=1)
```

```python
from sklearn.metrics import r2_score
import tensorflow as tf

def r2_metric(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    return r2
```

```python
import json
import numpy as np
def convert_to_python_type(obj):
    if isinstance(obj, np.float32):
        return float(obj)
    raise TypeError

# Assuming you have the array of prices and the JSON objects in a list
def write_test_json(prices, out_filename):
    input_file_path = 'test.json'
    with open(input_file_path, 'r') as input_file:
        json_objects = json.load(input_file)    
    # Iterate through the JSON objects and assign prices
    for i, json_obj in enumerate(json_objects):
        json_obj['pret'] = prices[i]
    
    print(json_objects[0])
    # Save the updated JSON objects to a new file
    with open(out_filename, 'w') as output_file:
        json.dump(json_objects, output_file, indent=4, default=convert_to_python_type)
    
    print(f"New JSON file created: {out_filename}")
```

# PREPROCESS

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

X_train = X_train.drop('marca', axis=1)
X_train = X_train.drop('model', axis=1)
X_train = pd.get_dummies(X_train, columns=["cutie_de_viteze", "combustibil", "transmisie", "caroserie", "culoare"])

scaler = MinMaxScaler()
X_train[["an", "km", "putere", "capacitate_cilindrica"]] = scaler.fit_transform(X_train[["an", "km", "putere", "capacitate_cilindrica"]])

current_year = 2024
X_train['age'] = current_year - X_train['an']

X_train = X_train.drop(["optiuni_culoare", "id"], axis=1)

X_train['addons_count'] = X_train['addons'].apply(len)

all_addons = set(addon for addons_list in X_train['addons'] for addon in addons_list)

for addon in all_addons:
    X_train[addon] = X_train['addons'].apply(lambda addons_list: 1 if addon in addons_list else 0)

X_train = X_train.drop('addons', axis=1)

y_train = y_train.astype(float)
X_train = X_train.astype(float)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
X_test = X_test.drop('marca', axis=1)
X_test = X_test.drop('model', axis=1)
X_test = pd.get_dummies(X_test, columns=["cutie_de_viteze", "combustibil", "transmisie", "caroserie", "culoare"])

scaler = MinMaxScaler()
X_test[["an", "km", "putere", "capacitate_cilindrica"]] = scaler.fit_transform(X_test[["an", "km", "putere", "capacitate_cilindrica"]])

current_year = 2024
X_test['age'] = current_year - X_test['an']

X_test = X_test.drop(["optiuni_culoare", "id"], axis=1)

X_test['addons_count'] = X_test['addons'].apply(len)

# Extract unique addons
all_addons = set(addon for addons_list in X_test['addons'] for addon in addons_list)

# Create binary columns for each addon
for addon in all_addons:
    X_test[addon] = X_test['addons'].apply(lambda addons_list: 1 if addon in addons_list else 0)

# Drop the original 'addons' column
X_test = X_test.drop('addons', axis=1)

X_test = X_test.astype(float)
```

```python
missing_columns = set(X_train.columns) - set(X_test.columns)

# Add missing columns to X_test and fill with 0
for column in missing_columns:
    X_test[column] = 0

column_order = X_train.columns

# Reorder the columns in X_test based on the column order from X_train
X_test = X_test[column_order]
```

# NEURAL NETWORK

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score

# Define the models
gb_reg = GradientBoostingRegressor()
rf_reg = RandomForestRegressor()

# Define the parameter grids for grid search
gb_param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}

rf_param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Define the metrics you want to evaluate
scoring = {
    'neg_mean_squared_error': make_scorer(mean_squared_error, greater_is_better=False),
    'neg_mean_absolute_error': make_scorer(mean_absolute_error, greater_is_better=False),
    'r2': 'r2'
}

# Instantiate the grid search for Gradient Boosting Regression
grid_search_gb = GridSearchCV(estimator=gb_reg, param_grid=gb_param_grid, scoring=scoring, cv=5, refit='neg_mean_squared_error')
grid_search_rf = GridSearchCV(estimator=rf_reg, param_grid=rf_param_grid, scoring=scoring, cv=5, refit='neg_mean_squared_error')

# Fit the grid search to the data
grid_search_gb.fit(X_train, y_train)
grid_search_rf.fit(X_train, y_train)
# Fit the grid search to the data

print("Best parameters for Gradient Boosting Regression:", grid_search_gb.best_params_)
print("Best negative mean squared error:", grid_search_gb.best_score_)
print("Best negative mean absolute error:", grid_search_gb.cv_results_['mean_test_neg_mean_absolute_error'][grid_search_gb.best_index_])
print("Best R^2 score:", grid_search_gb.cv_results_['mean_test_r2'][grid_search_gb.best_index_])

print("\nBest parameters for Random Forest Regression:", grid_search_rf.best_params_)
print("Best negative mean squared error:", grid_search_rf.best_score_)
print("Best negative mean absolute error:", grid_search_rf.cv_results_['mean_test_neg_mean_absolute_error'][grid_search_rf.best_index_])
print("Best R^2 score:", grid_search_rf.cv_results_['mean_test_r2'][grid_search_rf.best_index_])
```

```python
best_gb_model = grid_search_gb.best_estimator_
best_rf_model = grid_search_rf.best_estimator_

# Assuming you have a separate test set (X_test, y_test)
# Predictions using Gradient Boosting
gb_predictions = best_gb_model.predict(X_val)

# Predictions using Random Forest
rf_predictions = best_rf_model.predict(X_val)

# Calculate R^2 score for both models
r2_gb = r2_score(y_val, gb_predictions)
r2_rf = r2_score(y_val, rf_predictions)

# Print the R^2 scores
print("R^2 score for Gradient Boosting on test set:", r2_gb)
print("R^2 score for Random Forest on test set:", r2_rf)
```
